```python
from tkinter import *

# Initialize global variables
board = [[0 for j in range(7)] for i in range(6)]
current_player = 1

# Initialize the game window
window = Tk()
window.title('Connect 4')
canvas = Canvas(window, width=700, height=600)
canvas.pack()

# Create the game board
for i in range(6):
    for j in range(7):
        canvas.create_rectangle(j*100+50, i*100+50, j*100+150, i*100+150, fill='yellow')

# Add a disc to the game board when a column is clicked
def add_disc(column):
    global board, current_player
    row = 5
    while row >= 0:
        if board[row][column] == 0:
            board[row][column] = current_player
            canvas.create_oval(column*100+55, row*100+55, column*100+145, row*100+145, fill='red' if current_player == 1 else 'blue')
            if check_win(current_player):
                print('Player', current_player, 'wins!')
                reset_game()
                return
            current_player = 2 if current_player == 1 else 1
            break
        row -= 1

# Check if a player has won the game
def check_win(player):
    global board
    for i in range(6):
        for j in range(4):
            if board[i][j] == player and board[i][j+1] == player and board[i][j+2] == player and board[i][j+3] == player:
                return True
    for i in range(3):
        for j in range(7):
            if board[i][j] == player and board[i+1][j] == player and board[i+2][j] == player and board[i+3][j] == player:
                return True
    for i in range(3):
        for j in range(4):
            if board[i][j] == player and board[i+1][j+1] == player and board[i+2][j+2] == player and board[i+3][j+3] == player:
                return True
    for i in range(3, 6):
        for j in range(4):
            if board[i][j] == player and board[i-1][j+1] == player and board[i-2][j+2] == player and board[i-3][j+3] == player:
                return True
    return False

# Reset the game board
def reset_game():
    global board, current_player
    board = [[0 for j in range(7)] for i in range(6)]
    current_player = 1
    canvas.delete('all')
    for i in range(6):
        for j in range(7):
            canvas.create_rectangle(j*100+50, i*100+50, j*100+150, i*100+150, fill='yellow')

# Bind the add_disc function to the canvas
canvas.bind('<Button-1>', lambda event: add_disc(event.x // 100))

# Start the game loop
window.mainloop()
```
